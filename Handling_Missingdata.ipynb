## Handling Missing Values
Real-world datasets often have missing data. Pandas offers:
- `isna()` / `isnull()` → detect missing values (returns True/False)
- `dropna()` → remove missing values (rows/columns)
- `fillna()` → replace missing values with a given value or method (e.g., forward fill)

Always inspect how much data is missing before deciding to drop or fill.

```python
import pandas as pd
import numpy as np
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
# isnull() or isna() detect missing values returns True
b = pd.DataFrame(data)
print(b.isna())
print(b.isna().sum())  # this will give count of null values row wise
print(b.dropna()) # it will remove entire row which causd lost in data
 # it will resolve the problem of dropna() method by keeping rows we need to passs value
# jo bhi int none  data type wala value ko mean sa replace kerta hai aur obj wali ko mode ya placeholder sa
# we can apply threshhold means drop a row only if it have at least X number of non-null value

```

```python
print(b)
b.dropna(thresh=4)
```

```python
# if we wantto fill null value manually then we use fillna
b.fillna(0)
# or we can pass custome value by targating cols
new_value = {
    'A': 0,
    'B': 1,
    'C': 2,
    'D': 3
}
b.fillna(value=new_value)
```

## Changing Data Types (`astype`)
- Convert a column to a different type using `astype()`.
- Examples: converting strings to numbers, integers to floats, or columns to category.

Be careful: make sure the data is compatible with the new type!

```python
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

c = pd.DataFrame(a)
c['Age'].astype(float) # this will do type conversion
# obj or str cannot typecase into int
```

## Replacing Values (`replace`)
- Use `replace()` to substitute specific values or patterns.
- Useful for cleaning labels, fixing typos, or mapping categories.

```python
c['Name'].replace('Emma','deep').astype(str)
c['Age'].replace(35,99).astype(str)
```

## Dropping Rows & Columns (`drop`)
- `drop(labels, axis=0)` → drop rows by index labels.
- `drop(labels, axis=1)` → drop columns by name.
- Use `inplace=True` if you want to modify the DataFrame directly.

```python
 #axis = 0 remove row
 # for axis = 1 remove col
b.drop(index = 2,axis =1)
```

## Detecting & Removing Duplicates
- `duplicated()` → returns a Boolean Series marking duplicate rows.
- `drop_duplicates()` → remove duplicate rows.
- You can specify `subset` (columns) and `keep` ('first', 'last', or False).

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "David"],
    "Age": [25, 30, 35, 25, 30, 25],
    "Score": [88, 92, 79, 88, 92, 85]
}

df = pd.DataFrame(data)
print(df.duplicated().sum()) # return number of duplicate
print(df.drop_duplicates(inplace = True)) # exect identical rows aati hai tho hee remove kerage
```
