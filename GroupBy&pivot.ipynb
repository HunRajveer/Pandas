# 🐼 Pandas - Class 6: GroupBy & Aggregations
Welcome to **Class 6** of our Pandas series. Today we’ll learn how to group data and perform aggregations.

## 1. Concept of Split–Apply–Combine
- **Split** the data into groups based on some criteria.
- **Apply** a function (e.g., sum, mean) to each group.
- **Combine** the results into a new DataFrame or Series.

This is the core idea behind `groupby`.

```python
import pandas as pd
import numpy as np

# New dataset: Employees, their Department, and Salary
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

## 2. Using `groupby()` with Aggregation Functions
- Use `groupby('col').mean()` or other functions like sum, count, min, max.
- You can also use `agg()` to pass multiple functions at once.

```python
# it si use to perform qualitative gropuing means based on catogary or region ect
df.groupby('Category')['Sales'] # output will be like (<pandas.core.groupby.generic.SeriesGroupBy object at 0x7edbbcc45a90>) meaens it will create an onject
print(df.groupby('Category')['Sales'].sum() )# always remember pass a col you need to group inside () then pass col which have numeric value then must pass agg fun or other opration
print(df.groupby('Category')['Sales'].median())
print(df.groupby('Category')['Sales'].mean())
print(df.groupby('Category')['Sales'].std())
#df.groupby('Category')['Sales'].mode() # it does not have mode
```

```python
# we can do multipe grouping
print(df.groupby(['Category', 'Store'])['Sales'].sum())
print(df.groupby(['Category', 'Store'])['Sales'].mean())
```

## 3. Multiple Aggregations on Different Columns
- With `agg()`, you can specify different functions for each column.
- Example: `df.groupby('Dept').agg({'Salary':'mean', 'Age':'max'})`.

## 4. Pivot Tables & Crosstab
- `pivot_table()` summarizes data like Excel pivot tables.
- `crosstab()` shows frequency counts of combinations of factors.
- Both are powerful for summarizing and comparing groups.

```python
from numpy._core.fromnumeric import mean
# pivot tables are customize table which is use to create a custom data from existing table

#format  pd.pivot_table(df,value by which we need to do analysis,index = row , column=,aggragation function )
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}
d = pd.DataFrame(data)
d
```

```python
pd.pivot_table(d,values='Sales',index ='Product',columns='Region',aggfunc='median')
# or we can also pass multiple value
pd.pivot_table(d,values=['Sales','Units'],index = 'Product',columns='Region',aggfunc='std') # defaault it will find mean for values
```
