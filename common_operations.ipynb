Welcome to **Class 2**. Today we’ll learn how to **inspect**, **select**, and **organize** data inside a DataFrame.

## 1 Inspecting Data
Quick ways to understand your dataset:
- `head(n)` → first *n* rows (default 5)
- `tail(n)` → last *n* rows
- `info()` → columns, non-null counts, and dtypes
- `shape` → (rows, columns)
- `dtypes` → data type of each column

```python
import pandas as pd

# Create a small DataFrame
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
print(b.head(3))# DEFAULT FIRST FIVE VALUE
# tail() this will return n rows
print(b.tail(2)) # same as head default 5 value from bottom
print(b.info()) # it will provide info like no of rows and col and datatype and null or notnull
print(b.shape)
print(b.dtypes) # dtypes will return data type of col
```

## 2 Selecting Columns & Rows
Ways to access parts of a DataFrame:
- Single column → `df['col']`
- Multiple columns → `df[['col1','col2']]`
- Label-based selection → `df.loc[row_label, col_label]`
- Integer-index selection → `df.iloc[row_idx, col_idx]`
Tip: `:` means "all rows/columns" in that dimension.

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
# we will do slecting cols
df[['Name','Age']] # we need to pass a list inside list when we want to select multiple cols

```

```python
# creating a new cols
df['experience']= ['12','10','31','9'] # make sure to enter data same langth
```

## 3 Index & Columns Overview
- `df.index` → row labels
- `df.columns` → column names
- `df.values` → underlying NumPy array (read-only; prefer using DataFrame ops)

```python
# b.index it will tell from where our index start and stop
print(df.index) # output - RangeIndex(start=0, stop=4, step=1)
#b.columns it will gives the name of cols
print(df.columns) # output- Index(['Name', 'Age', 'City', 'Salary', 'experience'], dtype='object')


#df.values # output -array([['John', 28, 'New York', 65000, '12'],
      # ['Anna', 34, 'Paris', 70000, '10'],
      # ['Peter', 29, 'Berlin', 62000, '31'],
      # ['Linda', 42, 'London', 85000, '9']], dtype=object)
print(df.values) #our dataframe will convert into matrix form and display
```

## 4 Renaming Columns & Setting an Index
- Rename columns → `df.rename(columns={'old':'new'}, inplace=False)`
- Set a column as index → `df.set_index('column_name', inplace=False)`
- Reset index → `df.reset_index(inplace=False)`
 If you don't use `inplace=True`, remember to assign the result back to a variable.

```python
# df.rename(columns={'old':'new'},inplace = ture) inplace will
# inplace will change original dataframe if its fales then no change else change
df.rename(columns={'Name':'Employee_Nme','Salary':'Income'},inplace = True)

# b.set_index('column_name', inplace=False) this will change the index


# we can reset also



```
