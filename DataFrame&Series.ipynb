## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
import pandas as pd # series is vertically stack and only have one col
marks = pd.Series([30,40,50,60,70],
                  index=['Ronaldo','Messi','Virat','Dhoni','Mbappe'])
print(marks)
````

### Creating a DataFrame

````python
#Multiple columns -> DataFrame. # data frame is a collection on series which allign vertically with sae langth
# we can create a series using list or array or dict
import numpy as np

data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
cols = ["Name","Age","City","Salary"]
arr = np.array(data_list)
# we need to pass col name becasuse default
df = pd.DataFrame(arr,columns= cols)
df
# this is how we create a dataframe using vector
````

### Creating DataFrame from NumPy Arrays

````python
information = pd.DataFrame([
  ['Rahul', 52, 80],
  ['Devraj', 22, 85],
  ['Abhishek',60,55]
], columns=['Name','Age','Marks'])

print(information)
print(information.loc[2])
print(information.loc[2, 'Marks'])
````

````python
#Generate a 3X4 matrix with random numbers (1-13) using numpy
#name the columns as [A,B,C,D]

import numpy as np
import pandas as pd

arr = np.random.randint(1,100, size=(3,4))

arr_df = pd.DataFrame(arr, columns=['A','B','C','D'])
print(arr_df)
````

````python
# creating dataframe using dict
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
````
