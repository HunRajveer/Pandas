```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Countries.csv")
df
```

```python
# First step alwayas remember is to preprocess data
# find null values and type , shape etc
df.shape
df.info()
```

```python
# which country has highest population
df[['country','population']].max()
```

```python
# what is the capital of a country with highest population
df[['capital_city','population']].max()
```

```python
# which country have least population
df[['country','population']].min()
```

```python
# top 5 countries with highest democratic score
df.nlargest(5,'democracy_score')
# another approch

df['democracy_score'].nlargest(5)
```

```python
# how many country lies in  Eastern Europe region
CountryOfEastEurope = df[df['region']=='Eastern Europe']
# we can find lenght of dataframe just by usin len(value) in pandas
len(CountryOfEastEurope)
```

```python
# find  2nd highest populated and lowest  agricultural_land
df.sort_values(by=['population','agricultural_land'],ascending=[True,False]).head(2)
```
