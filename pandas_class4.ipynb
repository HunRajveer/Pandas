# 🐼 Pandas - Class 4: Data Selection & Filtering
Welcome to **Class 4** of our Pandas series. Today we’ll explore techniques to select and filter data in a DataFrame.

## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

```python
import pandas as pd


data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [24, 30, 22, 35],
        'Score': [88, 92, 75, 80]}
df = pd.DataFrame(data)

# Simple Boolean Indexing: Select rows where Age > 25
older_than_25 = df[df['Age'] > 25]
print("Rows where Age > 25:")
print(older_than_25)

#  Select rows where Score is exactly 75
score_75 = df[df['Score'] == 75]
print("\nRows where Score is 75:")
print(score_75)
```

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

```python
# 1. Multiple conditions using AND (&)

condition_and = df[(df['Age'] > 20) & (df['Score'] > 80)]
print("Rows with Age > 20 AND Score > 80:")
print(condition_and)

#  Multiple conditions using OR (|)
#
condition_or = df[(df['Age'] < 25) | (df['Score'] > 90)]
print("\nRows with Age < 25 OR Score > 90:")
print(condition_or)
```

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

```python
# 1. Basic query using a string expression

result_query = df.query('Age > 25')
print("Query (Age > 25):")
print(result_query)

# 2. Complex query with multiple conditions

complex_query = df.query('Age < 30 and Score > 80')
print("\nComplex Query (Age < 30 and Score > 80):")
print(complex_query)

# 3. Using variables with the @ symbol
min_score = 90
var_query = df.query('Score >= @min_score')
print(f"\nQuery using variable (@min_score={min_score}):")
print(var_query)
```

## 4. String Operations on Columns
- Use the `.str` accessor for vectorized string methods.
- Common methods: `contains`, `replace`, `startswith`, `endswith`, `lower`, `upper`.
- Useful for cleaning or filtering text data.

```python
# 1. Using .str.contains() to filter rows
# Select rows where 'Name' contains the letter 'a' (case insensitive)
names_with_a = df[df['Name'].str.contains('a', case=False)]
print("Names containing 'a':")
print(names_with_a)

# 2. Using .str.upper() and .str.lower()
# Create new columns with different casing
df['Name_Upper'] = df['Name'].str.upper()
df['Name_Lower'] = df['Name'].str.lower()
print("\nDataFrame with Upper/Lower case names:")
print(df[['Name', 'Name_Upper', 'Name_Lower']])

# 3. Using .str.replace()
# Replace 'Alice' with 'Alison'
df['Name'] = df['Name'].str.replace('Alice', 'Alison')
print("\nDataFrame after replacing 'Alice' with 'Alison':")
print(df)
```
